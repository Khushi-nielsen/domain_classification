```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report 
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('data-2.csv')
```

```python
def conversion(file_path, output_path,column_to_convert):
    df=pd.read_csv(file_path)
    df = df.dropna() 
    df = df[df.astype(str).apply(lambda x: x.str.strip() != '').all(axis=1)]
    for column in column_to_convert:
        if column in df.columns:
            df[column]=df[column].str.lower().str.strip()
            mapping={'yes':0 , 'no':1}
            df[column]=df[column].map(mapping).astype('Int64')
    df.to_csv(output_path, index=False)
```

```python
input_file=output_file='data-2.csv'
column_to_convert=['status']
conversion(input_file,output_file,column_to_convert)
```

```python
df=pd.read_csv('data-2.csv')
```

```python
df.head()
```

```python
X=df['dnsdomains']
y=df['Status']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42,stratify=y)
```

```python
print(X_train.shape, X_test.shape,X.shape)
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=200)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
classifier = LogisticRegression(max_iter=200)
classifier.fit(X_train_tfidf, y_train)
```

```python
y_pred = classifier.predict(X_test_tfidf)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
y_pred_train = classifier.predict(X_train_tfidf)
y_pred_test = classifier.predict(X_test_tfidf)
```

```python
training_data_accuracy = accuracy_score(y_train, y_pred_train)
test_data_accuracy = accuracy_score(y_test, y_pred_test)
print("Accuracy on training data:", training_data_accuracy)
print("Accuracy on test data:", test_data_accuracy)
```
