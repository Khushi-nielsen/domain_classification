```python
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier
import boto3
from datetime import datetime 
import os
```

```python
bucket_name="nonprod-ubo-s3-useast1-dev"
```

```python
s3=boto3.client('s3')
s3.download_file(bucket_name,'gsrs-dev/model/model.pkl','model.pkl')
s3.download_file(bucket_name,'gsrs-dev/model/vectorizer.pkl','vectorizer.pkl')
```

```python
model=joblib.load("model.pkl")
vectorizer=joblib.load("vectorizer.pkl")
```

```python
s3.download_file(bucket_name,'gsrs-dev/daily_data/day1.csv','day1.csv')
df_new=pd.read_csv("day1.csv")
```

```python
X_new=vectorizer.transform(df_new['dnsdomains'])
predictions=model.predict(X_new)
df_new['prediction']=predictions
```

```python
today_str=datetime.now().strftime("%Y-%m-%d")
output_filename=f'predictions/predictions_{today_str}.csv'
df_new.to_csv(output_filename,index=False)
```

```python
s3.upload_file(output_filename,bucket_name,f'gsrs-dev/predictions/predictions_{today_str}.csv')
```
